```python
def process_data(df):
    pivot = df.pivot_table(index=['name', 'year'],columns='state', values='count')
    pivot.fillna(0.0, inplace=True)
    pivot['Nation'] = pivot.sum(axis=1)
    for section in bigRegions:
        pivot[section] = pivot[bigRegions[section]].sum(axis=1)   
    pivot = pivot[bigRegions.keys()+['Nation']]
    for year in range(1910,2018):
        pivot.loc(axis=0)[:,year] /= pivot.loc(axis=0)[:,year].sum()
    
    unstacked = pivot.unstack()
    unstacked.fillna(0.0, inplace=True)
    unstacked = unstacked[unstacked.max(axis=1)>.001]
    
    trend = unstacked['Nation'].T - unstacked['Nation'].T.shift(1)        
    data = unstacked.stack()
    data['Trendiness'] = trend.unstack() * 3
    return data
```

```python
def dist(a,b):
    return sum(abs(a - b))

def getDistPath(a,b,distFunction = dist, shiftPenalty=.03):
    
    backsteps = np.array([[0,-1],[-1,0],[-1,-1]])
#     return diffFunction(x,y)
    m = np.zeros([len(a)+1, len(b)+1]) #additional row and col for left/top border
    p = np.zeros_like(m)
    p[0] = 0
    p[:,0] = 1
    p[0,0] = -1
    for row in range(1,len(a)+1):
        for col in range(1,len(b)+1):
            
            right = m[row,col-1] + shiftPenalty
            down = m[row-1,col] + shiftPenalty
            diag = m[row-1,col-1] + distFunction(a[row-1], b[col-1])
            m[row,col] = min([right, down, diag])
            p[row,col] = np.argmin([right, down, diag])
   
    traceRow, traceCol = len(a), len(b)
    trace = []
    while [traceRow, traceCol] != [0,0]:
        trace = [[traceRow, traceCol]]+trace
        shift = backsteps[int(p[traceRow, traceCol])]
        traceRow += shift[0]
        traceCol += shift[1]
    trace = [[traceRow, traceCol]]+trace    
    return m[-1,-1], trace
        
    

    
```
